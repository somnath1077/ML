```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X1 = np.arange(0, 100, 1)
X2 = [0.85 * x + 1.6 + np.random.normal(0, 16, 1) for x in X1]
X = np.c_[X1, X2]
```

```python
mu = np.mean(X, axis=0).reshape(1, 2)
X_prime = X - mu
```

```python
plt.plot(X[:, 0], X[:, 1], linestyle='dotted')
plt.xlabel('Population in Thousands')
plt.ylabel('Ad Spending')
plt.show()
```

# PCA

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=1)
pca.fit(X)
```

```python
print(pca.explained_variance_ratio_) 
```

```python
X_trans = pca.transform(X)
```
