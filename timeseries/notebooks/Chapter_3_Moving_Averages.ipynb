```python
x = df_housing.index
y = df_housing.total_sales

yrs = np.sort(pd.unique(df_housing.year))

plt.figure(figsize=(16,8))
plt.xticks(np.arange(5, len(df_housing), step=12), yrs)
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.plot(x, y)
plt.show()
```
